```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('housing.csv')
df.head()
```

```python
x = df['median_income']
y = df['median_house_value']

x_mean = np.mean(x)
y_mean = np.mean(y)

numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)
slope = numerator / denominator

intercept = y_mean - slope * x_mean

y_pred = intercept + slope * x

residual_error = y - y_pred

mae = np.mean(np.abs(residual_error))

mse = np.mean(residual_error ** 2)

rmse = np.sqrt(mse)
```

```python
print('Mean of X (median_income):', x_mean)
print('Mean of Y (median_house_value):', y_mean)
print('Slope:', slope)
print('Intercept:', intercept)
print('First 5 predicted values:', y_pred.head().tolist())
print('First 5 residual errors:', residual_error.head().tolist())
print('MAE:', mae)
print('MSE:', mse)
print('RMSE:', rmse)
```
